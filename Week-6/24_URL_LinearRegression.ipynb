```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
url="https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv"
data=pd.read_csv(url)
print("Dataset Preview")
print(data.head())
X=data["Experience Years"]
Y=data["Salary"]
model=LinearRegression()
model.fit(X.values.reshape(-1,1),Y)
print("\nModel Parameters")
print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)
years=[[5]]
predicted_salary=model.predict(years)
print("\nPredicted Salary for 5 years of experience:",predicted_salary[0])
y_pred=model.predict(X.values.reshape(-1,1))
mse=mean_squared_error(Y,y_pred)
print("\nMean Squared Error:",mse)
plt.scatter(X,Y)
plt.plot(X,y_pred)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear regression salary prediction")
plt.show()
```
