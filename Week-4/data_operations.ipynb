```python
import sqlite3
conn=sqlite3.connect("example.db")
cursor=conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    email TEXT UNIQUE
)
""")
print("Table created successfully.")
cursor.execute("INSERT INTO users(name, age, email) VALUES('Owaiz Khan',22,'owaizkhan117@gmail.com')")
cursor.execute("INSERT INTO users(name, age, email) VALUES('Ayesha',24,'ayesha230406@gmail.com')")
conn.commit()
print("Data inserted successfully.")
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()
print("\nUsers in database:")
for row in rows:
    print(row)
cursor.execute("UPDATE users SET age = 25 WHERE name = 'Owaiz'")
conn.commit()
print("\nRecord updated successfully.")
cursor.execute("DELETE FROM users WHERE name = 'Ayesha'")
conn.commit()
print("Record deleted successfully.")
cursor.execute("SELECT * FROM users")
rows=cursor.fetchall()
print("\nFinal data:")
for row in rows:
    print(row)
conn.close()
print("\nDatabase connection closed.")
```
