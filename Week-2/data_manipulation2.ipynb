```python
#writing data in pandas
import pandas as pd
data = {
    "Name" : ['Abdullah','Rahul','Jai','Salman'],
    "Age" : [20,21,22,19],
    "Marks" : [70,85,75,90]
}
df = pd.DataFrame(data)
print('Original data:- ')
print(df)

#Filtering Data
print("Alice's data:- ")
print(df[df["Name"] == 'Alice'])
print("Students with marks less than 85:- ")
print(df[df["Marks"] < 85])
print("Student with marks greater than 75 and age less than 20:- ")
print(df[(df["Marks"] > 75) & (df["Age"] < 20)])
```

```python
#Data Manipulation
import pandas as pd
data = {
    "Name" : ['Abdullah','Rahul','Jai','Salman'],
    "Age" : [20,21,22,19],
    "Marks" : [70,85,75,90]
}
df = pd.DataFrame(data)
print('Original data:- ')
print(df)
#Adding a new column
df['Bonus'] = 5
print('Data after adding a new column:- ')
print(df)
#Increasing marks
df['Marks'] = df['Marks'] + 2
print("After increasing marks:- \n")
print(df)
#Create new column from existing columns
df["Total"] = df["Marks"] + df["Bonus"]
print("After creating new column:- \n")
print(df)
#Renaming a column
df.rename(columns={"Total" : "Final"},inplace = True)
print("After renaming:- \n")
print(df)
#Removing a column
df.drop("Bonus",axis=1,inplace=True)
print("After removing a column:- \n")
print(df)
```
