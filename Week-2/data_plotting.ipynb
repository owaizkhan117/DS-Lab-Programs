```python
#Line Graph
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5,10,15,20,25]
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("x-y graph")
plt.show()
```

```python
#Bar Graph
import matplotlib.pyplot as plt
subjects=['Accounting','Statistics','Sales','Probability']
marks=[95,85,74,62]
plt.bar(subjects,marks)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks in subjects")
plt.show()
```

```python
#Scattered Graph
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[55,75,85,65,95]
plt.scatter(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter graph")
plt.show()
```

```python
#Pie Chart
import matplotlib.pyplot as plt
activities=['Code','Sleep','Eat','Drink','Smoke']
hours=[6,8,2,2,1]
plt.pie(hours,labels=activities,autopct="%1.1f%%")
plt.title("Daily activities")
plt.show()
```
