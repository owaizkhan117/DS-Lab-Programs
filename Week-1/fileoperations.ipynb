```python
'''4.Write the code to create, save and read the
   file from the local drive
   to perform the text operations ( count number of lines, words and characters)'''
#Create and write data into a file
file_name = "sample_text.txt"
with open (file_name,"w")as f:
  f.write("Hello World\n")
  f.write("This is a file created using python written in colab")
  f.write("Counting lines words and characters")

print("File created and data written!")

# Read file and perform operations
with open(file_name, "r") as f:
    text = f.read()

# Count lines
lines = text.split("\n")
line_count = len(lines)

# Count words
word_count = len(text.split())

# Count characters
char_count = len(text)

print("Number of lines:", line_count)
print("Number of words:", word_count)
print("Number of characters:", char_count)

```
