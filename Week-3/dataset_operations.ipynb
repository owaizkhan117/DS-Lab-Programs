```python
#13. Implementation of mathematical operations on pandas dataframe
import pandas as pd
data = {
    "Ages" : [22, 25, 47, 35, 46, 50, 28, 33, 39, 41],
    "Salaries": [3500, 3250, 5500, 4050, 5200, 6500, 3550, 3750, 4520, 4800],
}
df=pd.DataFrame(data)
print("Mean of the data is - \n",df.mean())
print("\n")
print("Median of the data is - \n",df.median())
print("\n")
print("Mode of the data is - \n",df.mode())
print("\n")
print("Standard deviation of the data is - \n",df.std())
print("\n")
print("Kurtosis of the data is - \n",df.kurt())
print("\n")
print("Skewness of the data is - \n",df.skew())
```

```python
import pandas as pd

url = "https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv"
data = pd.read_csv(url)
print("Origial Data:- \n")
print(data.head())
print("\n")

#Data Filtering
filtered_data = data[data[' Height(Inches)"']>70]
print("Filtered Data:- \n")
print(filtered_data)
print("\n")

#Data Mnipulation
data['Height_cm'] = data[' Height(Inches)"'] * 2.54
print("Insertion of new column (Data Manipulation):- \n")
print(data.head())
print("\n")

#Data Ranking
data['Weight_rank'] = data[' "Weight(Pounds)"'].rank(ascending=False)
print("Ranked Data:- \n")
print(data.sort_values('Weight_rank').head())
print("\n")

```
